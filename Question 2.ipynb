```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rcParams

rcParams['figure.figsize'] = (9, 6)
rcParams['lines.linewidth'] = 0.1
rcParams['font.size'] = 13
```

```python
week = ['M', 'T', 'W', 'Th', 'F']
b = [[3, 8.5, 11.5, 15, 1.3], [9.5, 5, 17, 4.5, 12.5], [21, 4, 1, 6, 2], 
     [2.5, 0.8, 20, 15, 14], [5, 2, 6, 4.5, 7]]
```

```python
df = pd.DataFrame(b, index = week, columns=['A', 'B', 'C', 'D', 'E'])
```

```python
df.plot(kind='bar',  alpha = 0.6, stacked=True, title = "user activity", rot ='horizontal')
```
